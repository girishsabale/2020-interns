```python
data = pd.read_json("data.json")
latest = pd.read_json("latest-rates.json")
data_f = pd.DataFrame(data)
print(data_f)
```

```python
#T-1 it will compare INR vs EUR from 1st jan-31st jan 2019

day = " "

INR = []
date = []

dt = 1

for dt in range(1,32):
    if(dt < 10):
        day = '0' + str(dt)
        print(day)
    else:
        day = str(dt)
        print(day)
        
    try:
        inr = data_f['rates']['2019-01-'+day]['INR']
        INR.append(inr)
        date.append('2019-01-'+day)
    except:
        print("An exception occurred")
print("\n")
print("Dates:-")
for i in date:
    print(i)        
print("\n")
print("1 Euro to INR:-")
for i in INR:
    print(i)
    
```

```python
#T-1 graph INR vs EUR 1st jan -31st jan 2019
x = date
y = INR

plt.figure(figsize=(10,10), facecolor="pink")
plt.xticks(rotation=90)

plt.xlabel("<---Dates--->")
plt.ylabel("<---INR Rate--->")

plt.title("INR Vs EUR from 2019:01:01 to 2019:01:31")

plt.plot(x,y,marker="^",label="INR",color="black")

plt.legend()
plt.show()
```
